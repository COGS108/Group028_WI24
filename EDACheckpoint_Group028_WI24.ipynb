# Research Question

Our research question sets out to examine what characteristics of movies influence their user rating score on IMDB. In particular, we will look at the runtime, budget, genre, MPA (Motion Picture Association) Rating, and release month to see if any of these metrics correlate with movie’s average user rating score on IMDB. In addition, we want to treat revenue as a secondary outcome variable, and see how each factor listed earlier correlates with a film's revenue. Furthermore, we are curious about the relationship between revenue made by the movies and its user ratings.

## Dataset #1: "Movie Industry" Dataset
**SETUP**

First, we install pandas and seaborn, which we will be using for our analysis.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.2f}'.format
```

Next, we load our dataset and look at the first few rows.

```python
df = pd.read_csv('./assets/movies.csv')
df.head()
```

Thankfully, it does not appear that our data needs much cleaning, as it is already relatively standardized. The main things we will need to do are to remove missing values, extract the release month from the 'released' column, and adjust the budget and revenue information for inflation.

First, we will drop all rows containing NA values, or rows where the "budget", "runtime", or "gross" are equal to 0. In the dataset notes on Kaggle, it was specified that these numerical values have a 0 where there was no data available, so we are treating 0 as an NA for these columns. We are dropping the entire row if any column is NA, since we would not be able to compare it to the other films in the dataset.

```python
df = df.dropna()
df = df[(df['budget'] != 0) & (df['gross'] != 0) & (df['runtime'] != 0)]
```

Next, we will adjust our dataframe so it only includes the columns that are relevant to our research question. We chose to keep "year" even though it is not one of the factors we're interested in, since we will need to know what year a movie came out in to adjust the revenue and budget for inflation. 

Looking at our dataset now that we have removed the missing values and renamed some columns for clarity, we can see that there are 5421 observations for us to work with in our analysis. 

```python
#Grabbing relevant rows
df = df[['name', 'rating', 'genre', 'year', 'released', 'score', 'budget', 'gross', 'runtime']]
cleaned_df = df.rename(columns={'gross': 'revenue'})
cleaned_df
```

Next, we will be extracting the release month from the 'released' column.

To do so, we will be defining a function 'extract_month()' that checks the value of the 'released' column, and replaces it with only the month that is listed.

```python
def extract_month(string):
    
    #convert the strings into usable format
    string = string.lower()
    string = string.strip()

    #Check to see which month the release date information contains, then preserving only that month
    if 'january' in string:
        string = 'January'
    elif 'february' in string:
        string = 'February'
    elif 'march' in string:
        string = 'March'
    elif 'april' in string:
        string = 'April'
    elif 'may' in string:
        string = 'May'
    elif 'june' in string:
        string = 'June'
    elif 'july' in string:
        string = 'July'
    elif 'august' in string:
        string = 'August'
    elif 'september' in string:
        string = 'September'
    elif 'october' in string:
        string = 'October'
    elif 'november' in string:
        string = 'November'
    elif 'december' in string:
        string = 'December'
    
    # If there is no release month, drop it from the dataset.

    else:
        string = np.NAN

    return string
```

Next, we apply extract_month() to our 'released' column to ensure that it properly extracted the release month.

```python
cleaned_df['released'] = cleaned_df['released'].apply(extract_month)
```

Finally, we rename the 'released' column to reflect the fact that it now only stores the month that the film was released in.

```python
df = cleaned_df.rename(columns={'released' : 'month'})
```

## Dataset #2: "Consumer Price Index for All Urban Consumers"
The rest of our data cleaning involves adjusting the budget and revenue information for inflation. To do this, we will need to bring in a second dataset containing the Consumer Price Index in the US per year. This information is necessary for adjusting dollar amounts for inflation over time.

First, we import the CPI dataset and take a look at it.

```python
cpi = pd.read_csv('./assets/CPIAUCSL.csv')
cpi
```

First, we will convert the 'DATE' column to datetime objects for ease of use. Then, since our movies dataframe only contains films from 1980 to 2020, we can truncate our cpi dataframe to only contain the information relevant to those years. 

```python
cpi['DATE'] = pd.to_datetime(cpi['DATE'])
# Filtering cpi to only include the years 1980-2020
cpi = cpi[(cpi['DATE'].dt.year <= 2020) & (cpi['DATE'].dt.year >= 1980)]
```

This dataframe measures the CPI monthly. For our purposes, we will be averaging the CPI's of a given year into a single CPI value representing that year. This will give us an identical 'year' column between dataframes to merge them later on. We rename the 'DATE' column to 'year' to merge them, and 'CPIAUSCL' to 'CPI' for ease of use.

```python
# Group the CPI's by year, then take the mean of each year's CPI values
cpi = cpi.groupby(cpi['DATE'].dt.year).mean()
cpi.reset_index(drop=True)

# Truncate the information in the 'DATE' column to only show the year next to its calculated average CPI.
cpi['DATE'] = cpi['DATE'].dt.year
cpi = cpi.rename(columns={'DATE' : 'year', 'CPIAUCSL' : 'CPI'})
cpi.reset_index(drop=True)
```

Now that we have a table of CPI values for each year in our movies dataframe, we just need to calculate one more piece of information before we can merge our dataframes and begin to adjust our figures for inflation. 

To do so, you first set set a "base year" (in our case, 2020). Then, you divide the CPI in each year by the base year's CPI. Finally, you divide the actual dollar amount from the original year by that number to give you a figure adjusted for inflation to your base year. 

We will now add a new column to the CPI dataframe which stores the result of dividing each price index by the base year's CPI. We will call this value the 'inflation_adjustment'

```python
# Dividing each CPI value by the CPI value for 2020 and storing in a new column
cpi['inflation_adjustment'] = cpi['CPI'] / cpi.iloc[40, 1]
cpi
```

Now that we have all the information we will need in the cpi dataframe, we can merge our two dataframes on their shared 'year' column and adjust our budgets and revenues for inflation.

```python
# Merging our two dataframes using their shared 'year' column.

combined_df = df.merge(cpi, on='year')
combined_df
```

To adjust our budgets and revenues for inflation, we will divide each value by its inflation_adjustment and store the results in new columns, called 'adjusted_budget' and 'adjusted_revenue'. We will then print the dataset to see if our adjustments were successful. Finally, we keep only the columns needed for the analysis.

```python
combined_df['adjusted_budget'] = combined_df['budget'] / combined_df['inflation_adjustment']
combined_df['adjusted_revenue'] = combined_df['revenue'] / combined_df['inflation_adjustment']
combined_df.drop(columns=['budget', 'revenue', 'CPI', 'inflation_adjustment'], inplace=True)
combined_df
```

We now have a combined dataset showing each film, along with its characteristic information of interest to our research question, with its budget and revenue corrected for inflation to allows to perform valid comparisons. With all this done, we are ready to begin exploring our data and seeing what correlations exist.
